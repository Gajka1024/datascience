```python
#Stworzenie listy dat rozpoczynając od 01.11.2019
date = pd.date_range('20191101', periods=6)
```

```python
#Stworzenie tablicy z losowymi wartościamy przypisując do nich daty jako indeksy wierszy
pd.DataFrame(np.random.randn(6, 4), index=date, columns=list('ABCD'))
```

```python
#Stworzenie dwóch tablic danych
s1 = pd.Series([1, 2, 3, 4, 5, 6], index=pd.date_range('20190102', periods=6))
df = pd.DataFrame(np.random.randn(6, 4), index=date, columns=list('ABCD'))
df
s1
```

```python
#Przypisanie do nowej kolumny F danych z tablicy s1
df['F'] = s1
df
```

```python
#Ustawienie wartości zero we wskazanej komórce (po etykiecie)
df.at[date[0], 'A'] = 0
df
```

```python
#Ustawienie wartości zero we wskazanej komórce (po pozycji)
df.iat[1, 1] = 0
df
```

```python
#Przekopiowanie tablicy 
df2 = df.copy()
# Ustawienie wartości ujemnych w miejscach, gdzie były większe od zera
df2[df2 > 0] = -df2
df2
```

```python
#Dodanie nowej kolumny zawierającej wartość 1 w dwóch pierwszych wierszach
df2.loc[date[0]:date[1], 'E'] = 1
df2
```

```python
#Wypisanie tabeli bez wierszy w których brakuje danych 
df2.dropna(how='any')
```

```python
#Uzupełnienie brakujących danych wartością 
df2.fillna(value=5)
```

```python
#Uzyskanie maski logicznej wypisującej wartość prawidziwą w miejscu brakujących danych
pd.isna(df2)
```

```python
df = pd.DataFrame({"id": [1, 2, 3, 4, 5, 6],
                   "raw_grade": ['a', 'b', 'b', 'a', 'a', 'e']})
df
```

```python
#Nadanie na nową kolumnę "grade" typu kategorii
df["grade"] = df["raw_grade"].astype("category")
df
```

```python
#Zmiana nazw w kolumnie odpowiadającej za kategorie
df["grade"] = df["grade"].cat.set_categories(["very bad", 
                                              "bad", "medium",
                                              "good", "very good"])
df
```

```python
#Grupowanie według kolumny kategorialnej wraz z ilością rekordów z danej kategorii
df.groupby("grade").size()
```

```python
#Stworzenie tablicy losowych danych z indeksem dat
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2019', periods=1000))
ts
```

```python
#Przypisanie do ts sumy skumulowanej 
ts = ts.cumsum()
ts
```

```python
#import pakietu odpowiadającego za diagramy
import matplotlib.pyplot as plt
```

```python
#Narysowanie wykresu
ts.plot()
```

```python
#Stworzenie tablicy z losowymi danymi
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index,
                  columns=['A', 'B', 'C', 'D'])
df
```

```python
#Przypisanie sumy skumulowanej 
df = df.cumsum()
#Narysowwanie wykresu
df.plot()
#Dodanie do wykresu legendy
plt.legend(loc='best')
```
